```
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=50, max_depth=50, min_samples_split =10)
regressor.fit(X_train_new, Y_train_label)
Y_pred = regressor.predict(X_test_new)
print(f"Score of training set: {regressor.score(X_train_new, Y_train_label)}")
print(f"Score of test set: {regressor.score(X_test_new, Y_test_label)}")
```
