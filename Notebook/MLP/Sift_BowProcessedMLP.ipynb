```
import numpy as np
import pickle
import sklearn
import os

from sklearn.neural_network import MLPClassifier
```

```
X_train_new = np.load("Processed_Sift&Bowfeatures_train.npy", allow_pickle=True)
X_test_new = np.load("Processed_Sift&Bowfeatures_test.npy", allow_pickle=True)
```

```
cd /
```

```
cd content/drive/MyDrive/Kaggle/Image_n_Labels/
```

```
Y_train = np.load("trainLabels.npy", allow_pickle=True)
Y_test = np.load("testLabels.npy", allow_pickle=True)
```

```
mlp50 = MLPClassifier(hidden_layer_sizes=50)
mlp50.fit(X_train_new, Y_train)
print(mlp50.score(X_test_new, Y_test))
```

```
mlp100 = MLPClassifier(hidden_layer_sizes=100)
mlp100.fit(X_train_new, Y_train)
print(mlp100.score(X_test_new, Y_test))
```

```
mlp200 = MLPClassifier(hidden_layer_sizes=200)
mlp200.fit(X_train_new, Y_train)
print(mlp200.score(X_test_new, Y_test))
```
