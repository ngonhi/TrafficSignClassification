```
from sklearn.metrics import classification_report
estimators = [1,10,50,100,500]
def train_model(estimators, trainData, trainLabels, testData, testLabels):
    for n in estimators:
        print(f'SVM WITH C = {n}')
        if os.path.isfile(f"Sift&BowPreprocess_SVM_c{str(n)}.pkl"):
            print("[INFO] loading classifier: SVM trained on ori images...")
            svm = pickle.load(open(f"Sift&BowPreprocess_SVM_c{str(n)}.pkl", 'rb'))
            print("[INFO] Classifer is loaded as instance ::rf::")
        else:
            print("[INFO] pre-trained classifier not found. \n Training Classifier MLP")
            svm = LinearSVC(C = n, class_weight="balanced")
            svm.fit(trainData,trainLabels)
            print("[INFO] Succefully trained the classsifier. \n Saving the classifier for further use")
            pickle.dump(svm, open(f"Sift&BowPreprocess_SVM_c{str(n)}.pkl", 'wb')) 
            print("[INFO] Classifier Saved")
            
        predictions = svm.predict(testData)
 
        # show a final classification report demonstrating the accuracy of the classifier
        print("EVALUATION ON TESTING DATA FOR" + str(n) + 'SVM')
        print(classification_report(testLabels, predictions))
        print(svm.score(trainData, trainLabels))
        print('-------------------------------------------------------')
```

```
train_model(estimators, X_train, Y_train, X_test, Y_test)
```
